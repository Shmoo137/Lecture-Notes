```python
import pickle

# Use colors as a dictionary
infile = open('D:\Repositories\lecture_notes_1st_version\colors\colors_dict.pkl','rb')
colors_dict = pickle.load(infile)
infile.close()

print(colors_dict)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

tone = 'dark' # or 'medium' or 'light'
i=0
for color in colors_dict:
  ax.plot(t, s+i/5, c=colors_dict[color][tone])
  i+=1
ax.grid()

plt.show()
```
