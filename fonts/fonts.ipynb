```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

path = '.\Hero New Medium.otf'
custom_font = fm.FontProperties(fname=path, size=15)
```

```python
![image](font_examples.png "Font examples")
```

```python
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, axs = plt.subplots(1,2)

i=0
for ax in axs:
  ax.plot(t, s+i)
  i+=2
  ax.grid()
  ax.set_ylim(0,4)

axs[0].text(1.5, 3.5, "(a)", fontsize=12, fontproperties=custom_font)
axs[1].text(1.5, 0.5, "(b)", fontsize=12, fontproperties=custom_font)

plt.show()
```
